```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv(r"F:\Car Performance Prediction\car prediction.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
#checking null values
dataset.isnull().any()
```

```python
#no need for label encoding caz all of them are numerical
#no need for onehotencoding as there is no textual or categorical data
```

```python
#splitting the values in x and y
x=dataset.iloc[:,1:8].values
y=dataset.iloc[:,0:1].values
```

```python
x.shape
```

```python
y.shape
```

```python
#testing the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(mlr,open("mileage.pkl","wb"))
```

```python
#predicting
y_pred=mlr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
x_test
```

```python
#checking accuracy of the model
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pred)
```
